```python
import os
import json
import pandas as pd

# Define the path to the transaction data
path = '/Users/ptsaravanan/Documents/DATA SCIENCE/Project/CAPSTONE Project/Phonepe Pulse DV/pulse/data/aggregated/transaction/country/india/state'
aggregated_state_list = [state for state in os.listdir(path) if os.path.isdir(os.path.join(path, state))]

# Initialize dictionary to store data for DataFrame creation
column = {
    'Transaction_type': [],
    'Transaction_count': [],
    'Transaction_amount': [],
    'State': [],
    'Year': [],
    'Quarter': []
}

# Check if the given path exists
if not os.path.exists(path):
    raise FileNotFoundError(f"The specified path does not exist: {path}")

# Iterate through state directories
for state in aggregated_state_list:
    state_path = os.path.join(path, state)
    year_list = [year for year in os.listdir(state_path) if os.path.isdir(os.path.join(state_path, year))]

    for year in year_list:
        year_path = os.path.join(state_path, year)
        files = os.listdir(year_path)

        for file in files:
            file_path = os.path.join(year_path, file)

            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    file_content = json.load(f)

                # Extract transaction data
                transaction_data = file_content.get("data", {}).get("transactionData", [])

                for transaction in transaction_data:
                    type_of_payment = transaction.get("name", "Unknown")
                    count = transaction.get("paymentInstruments", [])[0].get("count", 0)
                    amount = transaction.get("paymentInstruments", [])[0].get("amount", 0)

                    # Add data to the dictionary
                    column['Transaction_type'].append(type_of_payment)
                    column['Transaction_count'].append(count)
                    column['Transaction_amount'].append(amount)
                    column['State'].append(state)
                    column['Year'].append(year)
                    column['Quarter'].append(int(file.strip('.json')))

            except FileNotFoundError:
                print(f"File not found: {file_path}")
            except json.JSONDecodeError:
                print(f"Error decoding JSON in {file_path}")
            except Exception as e:
                print(f"An unexpected error occurred: {e}")

# Create DataFrame from the dictionary
df_aggregated_trans = pd.DataFrame(column)

df_aggregated_trans
```

```python
import os
import json
import pandas as pd

# Define the path to the user data
path_user = '/Users/ptsaravanan/Documents/DATA SCIENCE/Project/CAPSTONE Project/Phonepe Pulse DV/pulse/data/aggregated/user/country/india/state/'


# Initialize a dictionary to store data for DataFrame creation
data_dict = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Brand': [],
    'Count': [],
    'Percentage': []
}

# Get the list of state directories
states = [state for state in os.listdir(path_user) if os.path.isdir(os.path.join(path_user, state))]

for state in states:
    state_path = os.path.join(path_user, state)
    years = [year for year in os.listdir(state_path) if os.path.isdir(os.path.join(state_path, year))]

    for year in years:
        year_path = os.path.join(state_path, year)
        files = os.listdir(year_path)

        for file in files:
            file_path = os.path.join(year_path, file)

            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    file_content = json.load(f)

                # Extract usersByDevice data
                users_by_device = file_content.get("data", {}).get("usersByDevice", [])

                if not users_by_device:
                    print(f"No 'usersByDevice' found in {file_path}")
                    continue

                # Append data to the dictionary
                quarter = int(file.strip('.json'))

                for device in users_by_device:
                    brand = device.get("brand", "Unknown")
                    count = device.get("count", 0)
                    percentage = device.get("percentage", 0)

                    data_dict['Brand'].append(brand)
                    data_dict['Count'].append(count)
                    data_dict['Percentage'].append(percentage)
                    data_dict['State'].append(state)
                    data_dict['Year'].append(year)
                    data_dict['Quarter'].append(quarter)

            except FileNotFoundError:
                print(f"File not found: {file_path}")
            except json.JSONDecodeError:
                print(f"Error decoding JSON in {file_path}")
            except Exception as e:
                print(f"An unexpected error occurred: {e}")

# Create a DataFrame from the data dictionary
df_aggregated_user = pd.DataFrame(data_dict)
```

```python
import pandas as pd
from sqlalchemy import create_engine, Table, Column, String, Integer, Float, MetaData, text

engine = create_engine("mysql+mysqlconnector://root:Blue9791@localhost/")

# Create a new database if it doesn't exist
with engine.connect() as conn:
    conn.execute(text("CREATE DATABASE IF NOT EXISTS phonepe_pulse"))

# Create SQLAlchemy engine and connect to MySQL
engine = create_engine("mysql+mysqlconnector://root:Blue9791@localhost/phonepe_pulse")

# MetaData is used to hold information about tables and database
metadata = MetaData()

# Define table structures
agg_trans = Table('agg_trans', metadata,
                  Column('State', String(100)),
                  Column('Year', Integer),
                  Column('Quarter', Integer),
                  Column('Transaction_type', String(100)),
                  Column('Transaction_count', Integer),
                  Column('Transaction_amount', Float))

agg_user = Table('agg_user', metadata,
                 Column('State', String(100)),
                 Column('Year', Integer),
                 Column('Quarter', Integer),
                 Column('Brand', String(100)),
                 Column('Count', Integer),
                 Column('Percentage', Float))

# Continue with other tables in a similar fashion
map_trans = Table('map_trans', metadata,
                  Column('State', String(100)),
                  Column('Year', Integer),
                  Column('Quarter', Integer),
                  Column('District', String(100)),
                  Column('Count', Integer),
                  Column('Amount', Float))

map_user = Table('map_user', metadata,
                 Column('State', String(100)),
                 Column('Year', Integer),
                 Column('Quarter', Integer),
                 Column('District', String(100)),
                 Column('RegisteredUser', Integer),
                 Column('AppOpens', Integer))

top_trans = Table('top_trans', metadata,
                  Column('State', String(100)),
                  Column('Year', Integer),
                  Column('Quarter', Integer),
                  Column('Pincode', Integer),
                  Column('Transaction_count', Integer),
                  Column('Transaction_amount', Float))

top_user = Table('top_user', metadata,
                 Column('State', String(100)),
                 Column('Year', Integer),
                 Column('Quarter', Integer),
                 Column('Pincode', Integer),
                 Column('RegisteredUsers', Integer))

# Create tables in the database
metadata.create_all(engine)

# Load CSV files into the SQLAlchemy tables
agg_trans_df = pd.read_csv("agg_trans.csv")
agg_user_df = pd.read_csv("agg_user.csv")
map_trans_df = pd.read_csv("map_trans.csv")
map_user_df = pd.read_csv("map_user.csv")
top_trans_df = pd.read_csv("top_trans.csv")
top_user_df = pd.read_csv("top_user.csv")

# Use pandas to_sql to insert data into SQLAlchemy tables
agg_trans_df.to_sql('agg_trans', engine, if_exists='append', index=False)
agg_user_df.to_sql('agg_user', engine, if_exists='append', index=False)
map_trans_df.to_sql('map_trans', engine, if_exists='append', index=False)
map_user_df.to_sql('map_user', engine, if_exists='append', index=False)
top_trans_df.to_sql('top_trans', engine, if_exists='append', index=False)
top_user_df.to_sql('top_user', engine, if_exists='append', index=False)
```
